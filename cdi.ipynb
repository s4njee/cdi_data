```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
my_dpi = 137.68
sns.set(rc = {'figure.figsize':(18,8)})
```

```python
df = pd.read_csv(r'C:\Users\Sanjee\PycharmProjects\cdi\data\cdi.csv', low_memory=False)
```

```python
df.head()
```

```python
df.Question.unique()
```

```python
df = df[(df.Question == 'Influenza vaccination among noninstitutionalized adults aged >= 18 years')]
```

```python
df.head()
```

```python
df2 = df[(df.DataValueTypeID == 'AGEADJPREV') & (df.StratificationID1 == 'OVR')]
```

```python
df2.head()
```

```python
for year in range(2011,2020):
    dfyear = df2[df2.YearStart == year]
    dfyear.columns
    dfyear.DataValue = pd.to_numeric(dfyear.DataValue)
    dfyear.sort_values('DataValue', ascending=False, inplace=True)
    plt.figure()
    ax = sns.barplot(x='LocationAbbr', y='DataValue', data=dfyear)
    ax.set(xlabel='State', ylabel='Percent Vaccinated')
    ax.set_title(f'Influenza Vaccination Rates ({year})')
    plt.savefig(f'C:\\Users\\Sanjee\\PycharmProjects\\cdi\\overallflurate{year}.jpg')
    plt.clf()
```
